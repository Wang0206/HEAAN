```python
import numpy as np


with open('in.txt', 'r') as file:
    lines = file.readlines()
print(len(lines))

data = []
current_section = []
for line in lines:
    line = line.strip()
    if line == 'neg':
        continue
    elif line.startswith('c,c'):
        continue
    elif line.isdigit() or line.startswith('-'):
        current_section.append(int(line))
    elif line in ['finish conv', 'finish BN', 'finish RELU']:
        if current_section:
            data.append(current_section)
            current_section = []

if current_section:
    data.append(current_section)

i = 0
for sublist in data:
  print(len(sublist))
  i = i + 1

print(data[18])
```

```python

print("| 子列表 | 不同项个数 | 每个项重复次数 |")
print("| --- | --- | --- |")

for sub_list in data:
    unique_items = set(sub_list)
    item_counts = {}
    for item in sub_list:
        if item in item_counts:
            item_counts[item] += 1
        else:
            item_counts[item] = 1

    item_counts_str = ", ".join([f"{k}: {v}" for k, v in item_counts.items()])
    print(f"|  | {len(unique_items)} | {item_counts_str} |")
```
